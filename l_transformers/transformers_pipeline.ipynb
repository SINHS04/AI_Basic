```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
print(classifier("I've been waiting for a HuggingFace course my whole life."))
print(classifier(["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]))
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

```python
classifier = pipeline("sentiment-analysis", model="nlp04/korean_sentiment_analysis_kcelectra")
print(classifier("하,,,,내일 옥상다이브 하기 전에 표 구하길 기도해주세요."))
print(classifier("밴드 사운드 진짜 너무너무 좋다.... 꿈인가 진짜"))
print(classifier("칸 태리 너무 풋풋하고 귀여워.. 근데 허리가 한줌이셔 &others&"))
```
